# Intro

we used a dataset that recorded the top 23 crypto currencies and their prices across several years. prices were marked in a traditional stock exchange fashion where the beginning of the day at midnight is recorded as the opening price and the close price of each day is at 11:59:59 PM. This is not a truely effective way to record the market data of crypto because it is not traded in a physical exchange like the stock market but rather on a digital exchange that is running 24/7. Another thing to note is that each exchange can have a slightly different estimate for the current market value because their could be competing writes to the blockchain that could occur on the exact moment that an open or close occurs. This is rare and nearly imposible to identify with the dataset we have. Most of the issues we found when comparing this data against price histories on various exchanges was negligable and could be overlooked for what we wanted to examine. some exchanges had price differences of a few cents on smaller coins or even a dollar or two on big coins like ETH or BTC. Overall this data set was very accurate and we had a good amount of confidence in the accuracy.

Our main goal was to create classifiers to identify the bull and bear market trends shown in the data. Firstly let me clarify what bull and bear markets are. A bull market is when the current market price closes above the Exponential moving average (EMA), and a bear market is the inverse of that where the close in below the EMA. Bull markets are periods of growth in the valuation of the coin and point towards profitable trades at high values. Bear markets are periods of either stagnation where the value remains stable, or periods of decline in valuation. We created classifiers to keep track of these trends to later predict growth or decay in the price of each coin. This tool set can be used to save your investments before large market corrections and selloffs happen at the peaks after bull markets taper off, and inversley it can tell an investor when the coin has found a more stable position and gearing for another bull run so they can invest.

The source of this data set is www.Kaggle.com

# Visualizations and Feature Building
Here we examined some of the features of the data and created some of our own out of the existing data to be used to fit classifiers to.

```python
dirNameIn = 'Data/Crypto Price History/'
for fnameIn in os.listdir(dirNameIn):
    #print(fnameIn)
    if(fnameIn[0] != '.'):
        table = MyPyTable().load_from_file(dirNameIn+fnameIn)
        
        print("Coin:", fnameIn)
        DMAOpenCloseTable = utils.two_attr_difference_table(["Sample Number", "Close-Open"],table, "SNo", "Open", "Close", False)
        DMAHighLowTable = utils.two_attr_difference_table(["Sample Number", "High-Low"], table, "SNo", "Low", "High", False)
        five_day_SMA = utils.sma_table(["average number", "20 Day SMA"], table, "Close", 20)  
        volatility_table = utils.two_attr_difference_table(["Sample Number", "volatility"], table, "SNo", "Open", "Close", True)
        twenty_day_EMA = utils.ema_table(["ema#", "20 Day EMA"], table, "Close", 20)
        
        
        plot_utils.scatter(volatility_table, "Sample Number", "volatility")
        plot_utils.scatter(twenty_day_EMA, "ema#", "20 Day EMA")
        #plot_utils.scatter(five_day_SMA, "average number", "20 Day SMA")
        plot_utils.scatter(table, "SNo", "Close")
        #plot_utils.scatter(DMAOpenCloseTable, "Sample Number", "Close-Open")
        #plot_utils.scatter(DMAHighLowTable, "Sample Number", "High-Low")
        
        close_idx = table.column_names.index("Close")
        volume_quartiles = utils.calc_quartiles(sorted(table.get_column("Volume")))
        bullishvsbearish = ["Bullish" if table.data[i][close_idx] > twenty_day_EMA.get_column("20 Day EMA")[i] else "Bearish" for i in range(len(twenty_day_EMA.data))]
        gainsvslosesTmr = ["Gains" if table.data[i][close_idx] < table.data[i+5][close_idx] else "Loses" for i in range(len(table.data)-5)]
        percent_change_ratings = utils.percents_to_categorical(utils.percent_change_column(table.column_names, table, "Open", "Close"))
        volumeRatings = utils.volumes_to_categorical(table.get_column("Volume"), volume_quartiles)
        volatilityRatings = utils.split_on_quantiles(volatility_table.get_column("volatility"), utils.calc_quartiles(sorted(volatility_table.get_column("volatility"))))
        #print(volatilityRatings)
        
        header = ["Percent Change Rating", "Bullish or Bearish", "Volume Rating", "Volatility", "Gains or Loses Tomorrow"]
        changed_data = [[percent_change_ratings[i], bullishvsbearish[i], volumeRatings[i], volatilityRatings[i], gainsvslosesTmr[i]] for i in range(len(twenty_day_EMA.data))]
        changed_table = MyPyTable(header, changed_data)
        changed_table.save_to_file(dirNameIn+"../Updated/changed"+fnameIn)
        
        random_forest = MyRandomForestGenerator()
        
        data = [[changed_data[i][j] for j in range(len(changed_data[i])-1)] for i in range(len(changed_data))]
        X_train, X_test, y_train, y_test = myevaluation.train_test_split(data, changed_table.get_column("Gains or Loses Tomorrow"))
        
        
        
        random_forest.fit(X_train, y_train, 20, 7, 2)
        predictions = random_forest.predict(X_test)
        accuracy = sum([1 for i in range(len(predictions)) if predictions[i] == y_test[i]])/len(predictions)
        print("Random Forest Accuracy ", accuracy)
        conf_mat = myevaluation.confusion_matrix(y_test, predictions, list(set(y_test)))
        myutils.print_mat(list(set(y_test)), conf_mat)
        
        zeroR = MyRandomClassifer()
        
        zeroR.fit(X_train, y_train)
        zeroRPred = zeroR.predict(X_test)
        zeroRAcc = sum([1 for i in range(len(zeroRPred)) if zeroRPred[i] == y_test[i]])/len(zeroRPred)
        print("Random Classifier Accuracy:", zeroRAcc)
        conf_mat = myevaluation.confusion_matrix(y_test, zeroRPred, list(set(y_test)))
        myutils.print_mat(list(set(y_test)), conf_mat)
        
        model = MyDecisionTreeClassifier()
        
        model.fit(X_train, y_train)
        decision_tree_pred = model.predict(X_test)
        decTreeAcc = sum([1 for i in range(len(decision_tree_pred)) if decision_tree_pred[i] == y_test[i]])/len(decision_tree_pred)
        print("Decision Tree Classifier Accuracy:",decTreeAcc)
        conf_mat = myevaluation.confusion_matrix(y_test, decision_tree_pred, list(set(y_test)))
        myutils.print_mat(list(set(y_test)), conf_mat)
        
        model = MyKNeighborsClassifier()
        model.fit([[x if type(x) != str else 1 if x == "Bullish" else 0 for x in row] for row in X_train], y_train)
        lin_reg_pred = model.predict([[x if type(x) != str else 1 if x == "Bullish" else 0 for x in row] for row in X_test])
        linReg = sum([1 for i in range(len(lin_reg_pred)) if lin_reg_pred[i] == y_test[i]])/len(lin_reg_pred)
        print("K-Neared Neighbors Classifier Accuracy:", linReg)
        conf_mat = myevaluation.confusion_matrix(y_test, lin_reg_pred, list(set(y_test)))
        myutils.print_mat(list(set(y_test)), conf_mat)

        
```
